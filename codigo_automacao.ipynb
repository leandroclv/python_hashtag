Desafio:

Todos os dias, o nosso sistema atualiza as vendas do dia anterior. O seu trabalho diário, como analista, é enviar um e-mail para a diretoria, assim que começar a trabalhar, com o faturamento e a quantidade de produtos vendidos no dia anterior

E-mail da diretora: seugmail+diretoria@gmail.com
Local onde o sistema disponibiliza as vendas do dia anterior: https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga?usp=sharing

Para resolver isso, vamos usar o pyautogui, uma biblioteca de automação de comandos do mouse e do teclado

referência do pyautogui: https://pyautogui.readthedocs.io/en/latest/quickstart.html

Passo a passo do desafio

- Passo 1 -> Entrar no sistema da empresa (link do drive)
- Passo 2 -> Navegar no sistema para encontrar a base de dados 
- Passo 3 -> Exportar a base de dados (baixar o arquivo)

Passos 1, 2 e 3

```python
link = 'https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga?usp=sharing'
pyautogui.PAUSE = 1
pyautogui.press('win')
pyautogui.write('Vivaldi')
pyautogui.press('enter')
time.sleep(6)
pyautogui.click(x=1281, y=54)
pyautogui.write(link)
time.sleep(2)
pyautogui.press('enter')
time.sleep(8)
pyautogui.click(x=1169, y=267, clicks=2) #clicar no Exportar
time.sleep(10)
pyautogui.click(x=1199, y=269) #button='left' caso queira clicar com o botão direito do mouse/ clicar no arquivos vendas
time.sleep(10)
pyautogui.click(x=2137, y=271)# clicar no 'baixar'
time.sleep(10)
pyautogui.click(x=1514, y=281) #clicar no botão salvar
```

Passo 4

```python
caminho = r'/home/leandro/Downloads/Vendas - Dez.xlsx'
tabela = pd.read_excel(caminho)
```

```python
print(tabela)
```

```python
display(tabela)
```

```python
#somar o faturamento de todos os produtos = somar a coluna de Valor Final
#Somar a quantidade de produtos = somar a coluna de Quantidade

faturamento = tabela['Valor Final'].sum()
#print(f'O faturamento da empresa foi de R${faturamento:.2f} reais')

qutde_produtos = tabela['Quantidade'].sum()
#print(f'A quantidade de produtos vendidos foi de {qutde_produtos}')
```
