###Introdução a Orientação a Objetos

####Regras Gerais:
- Tudo no Python é objeto
    - String é objeto
    - Lista é objeto
    - Dicionários são objetos ...


####Comparação Clássica
- Pense no Controle Remoto de uma televisão
    - O controle é um objeto
    - Cada botão dele é um comando, um método
    - Cada método faz uma ação específica
        -Por trás de cada método (dentro do controle) podem acontecer milhares de coisas quando você aperta 1 botão, mas no fundo você tá cagando para isso, só quer que o botão faça o que você mandou quando você clicar no botão
