## 1. Faturamento do Melhor e do Pior Mês do Ano

Qual foi o valor de vendas do melhor mês do Ano?
E valor do pior mês do ano?

```python
meses = ['jan', 'fev', 'mar', 'abr', 'mai', 'jun', 'jul', 'ago', 'set', 'out', 'nov', 'dez']
vendas_1sem = [25000, 29000, 22200, 17750, 15870, 19900]
vendas_2sem = [19850, 20120, 17540, 15555, 49051, 9650]

vendas_total = vendas_1sem + vendas_2sem
maior = max(vendas_total)
print(maior)
menor = min(vendas_total)
print(menor)

#Vendas_1sem.extend(vendas_2sem)
#maior_valor = max(vendas_1sem)
#menor_valor = min(vendas_1sem)
#print(maior_valor)
#print(menor_valor)
```

## 2. Continuação

Agora relacione as duas listas para printar 'O melhor mês do ano foi {} com {} vendas' e o mesmo para o pior mês do ano.

Calcule também o faturamento total do Ano e quanto que o melhor mês representou do faturamento total.

Obs: Para o faturamento total, pode usar a função sum(lista) que soma todos os itens de uma lista

```python
i_maior = vendas_total.index(maior)
i_menor = vendas_total.index(menor)

print(f'O melhor mês do ano foi {meses[i_maior]} com {maior} vendas, já o pior mês de vendas foi {meses[i_menor]} com {menor} vendas.')

faturamento_ano = sum(vendas_total)
melhor_mes = maior / faturamento_ano
print(f'O faturamento total do ano foi de {faturamento_ano} é o mês de {meses[i_maior]} teve {melhor_mes * 100:.1f} do faturamento total')
```
