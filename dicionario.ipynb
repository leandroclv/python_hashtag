###Dicionário em Python

```python
#Estrutura:
#dicionário = {chave:valor, chave:valor, chave:valor, ...}
```

Vantagens e desvantagens:
- Não devem ser usados para pegar itens em uma determinada ordem
- Podem ter valores heterogêneos (vários tipos de valores dentro de um mesmo dicionário: inteiros, strings, listas, etc)
- Chaves são únicas obrigatoriamentes
- Mais intuitivos de trabalhar

```python
mais_vendidos = {'tecnologia': 'iphone', 'refrigeracao': 'ar consul 12000 btu', 'livros': 'o alquimista', 'eletrodoméstico': 'geladeira', 'lazer': 'prancha surf'}
vendas_tecnologia = {'iphone': 15000, 'samsung galaxy': 12000, 'tv samsung': 10000, 'ps5': 14300, 'tablet': 1720, 'ipad': 1000, 'tv philco': 2500, 'notebook hp': 1000, 'notebook dell': 17000, 'notebook asus': 2450}
```

```python
#Qual foi o item mais vendido nas categorias 'livros' e 'lazer'?
livro = mais_vendidos['livros']
lazer = mais_vendidos['lazer']
print(livro)
print(lazer)
```

```python
#Quanto foi vendido de 'notebook asus' e de 'ipad'?
qtde_notebook_asus = vendas_tecnologia['notebook asus']
qtde_ipad = vendas_tecnologia['ipad']
print(qtde_notebook_asus)
print(qtde_ipad)
```

###Não confie na ordem dos dicionários, use as chaves

Python versões antigas X versões novas
- Dicionários eram 'sem ordem'. Atualmente tem ordem, mas o certo é usar as chaves
- 2 formas de pegar um valor:
    - dicionário(chave)
    - get(chave)

```python
print(f'O livro mais vendido foi o {mais_vendidos['livros']}')
```

```python
vendas_tecnologia.get('notebook asus')
```

```python
print(f'Vendemos {vendas_tecnologia.get('ipad')} ipads')
```

```python
print(vendas_tecnologia.get('ipads'))
```

####Verificar se item está no dicionário:
- if
- get(chave) = None

```python
#Se tentarmos procurar as vendas de 'copo' na lista de vendas tecnologia, o que acontece?
if 'copo' in vendas_tecnologia:
    print(vendas_tecnologia['copo'])
else:
    print('Copo não está dentro da lista de produtos de tecnologias')
```

```python
if vendas_tecnologia.get('copo') == None:
    print('Copo não está dentro da lista de produtos de tecnologias')
else:
    print(vendas_tecnologia.get('copo'))
```

###Adicionar, remover e modificar itens no dicionário

#####Estrutura
- Adicinoar itens:
    - dicionario = {}
    - dicinoario[chave] = valor
    - outra opção:
    - dicionario.update({chave: valor, chave: valor})

```python
lucro_1tri = {'janeiro': 100000,'fevereiro': 120000, 'março': 90000}
lucro_2tri = {'abril': 88000, 'maio': 89000, 'junho': 120000}
```

```python
#Adicionando um item
print(lucro_1tri)
lucro_1tri['abril'] = 88000
print(lucro_1tri)
```

```python
#Adicinoando vários itens ou um dicionário a outro
lucro_1tri.update(lucro_2tri)
lucro_1tri
```

```python
#Modificar itens:
#Da mesma forma que adicionamos um valor, caso essa chave já exista o item é apenas modificado
#dicionario{chave} = valor
#vamor modificar o lucro de fevereiro (lembrando: caso o itm não exista, ele vai criar o item no dicionário)
```

```python
lucro_1tri['janeiro'] = 88000
lucro_1tri
```

```python
#Remover itens:
#del dicionario{chave}
#ou então
#valor = dicionario.pop(chave)
#mas cuidado com:
#del dicionario (deleta o dicionário) é diferente de dicionario.clear()(deleta todos os itens do dicionário)
```

```python
#remover o mês de junho
del lucro_1tri['junho']
print(lucro_1tri)
```

```python
lucro_janeiro = lucro_1tri.pop('janeiro')
print(lucro_1tri)
print(lucro_janeiro)
```

###Listas importantes a partir do Diconário

```python
chaves = vendas_tecnologia.keys()
chaves
```

###Transformar listas em dicionários e function zip

Estrutura:
    - Dicionário com valores padrões:

    dicionario = dict.fromkeys(lista_chaves, valor_padrao)

    - Dicnoário a partir de listas de tuplas:

    dicionario = dict(lista_tuplas)

    - Dicionário a partir de 2 listas:

    Passo 1: Transformar listas em lista de tuplas como o método zip

    Passo 2: Transformar em dicionário
    - lista_tuplas = zip(lista1, lista2)
    - dicionario = dict(lista_tuplas)

```python
produtos = ['iphone', 'samsung galaxy', 'tv samsung', 'ps5', 'tablet', 'ipad', 'tv philco', 'notebook hp', 'notebook dell', 'notebook asus']
vendas = [15000, 12000, 10000, 14300, 1720, 1000, 2500, 1000, 17000, 2450]
```

```python
dicionario = dict.fromkeys(produtos, 0)
dicionario
```

```python
lista_tuplas = zip(produtos, vendas) #zip -> junta listas
dicionario_vendas = dict(lista_tuplas)
print(dicionario_vendas)
```

```python
#Quanto vendemos de ipad?
indice = produtos.index('ipad')
indice
```

```python
vendas[indice]
```

```python
print(f'Vendemos {dicionario_vendas['ipad']} ipads')
```
