###Tipos de variáveis em Python
Uma variável é um objeto, afinal tudo é um objeto em Python

- int -> inteiro
- string -> texto
- float -> números com casas decimais (ponto flutuante)
- bool ou boolean -> True ou False

obs: Variáveis em Python não são declardas explicitamente, mas isso não significa que você deve ficar mudando o tempo todo

obs: Cuidado com os nomes restritos

```python
faturamento = 1000
type(faturamento)
```

```python
faturamento = 1000.00
type(faturamento)
```

```python
faturamento = '1000'
type(faturamento)
```

```python
ganha_bonus = True
type(ganha_bonus)
```

```python
faturamento = 2000
custo = 500
lucro = faturamento - custo

print('O faturamento da loja foi ' + str(faturamento))
print('O custo da loja foi ' + str(custo))
print('Assim, o lucro da loja foi de ' + str(lucro) + ' reais')
```

```python
faturamento = 2000
custo = 500
lucro = faturamento - custo
print('O faturamento da loja foi {}'.format(faturamento))
```

```python
#f-string
faturamento = 1000
custo = 400
lucro = faturamento - custo
print(f'O faturamento foi de {faturamento} e o lucro de {lucro}')
```
