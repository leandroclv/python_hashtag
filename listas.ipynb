###Listas em Python

####EStrutura:

lista = [valor, valor,valor, ...]

- Lista é um dos objetivos mais importantes de Python,por isso vamos trabalhar bastante neles
- Quando importamos uma base de dados para o Python, normalmente ele é lido como uma 'lista' ou como alguma 'variação de lista'
- Listas em Python foram feitas para serem homogêneas, apesar de aceitarem valores heterogêneos 
- Exemplos de lista:

```python
#Listas de produtos de uma loja:
produtos = ['tv', 'celular', 'mouse', 'teclado', 'tablet']

#nome = laptop
#produtos = ['tv', 'celular', 'mouse', 'teclado', 'tablet', nome]
#print(produtos) -> ['tv', 'celular', 'mouse', 'teclado', 'tablet', 'laptop']

#Lista de unidades vendidas de cada produto da loja:
vendas = [1000, 1500, 350, 270, 900]
```

```python
#Lista[i] -> é o valor de índice i da lista
#Obs: Lembre que no python o índice começa em 0, então o primeiro item de uma lista é o item lista[0]

#Para substituir um valor de uma lista você pode fazer:
#lista[i] = novo_valor
```

```python
print(f'Vendas do produto {produtos[0]} foram de {vendas[0]} unidades')
```

```python
vendas[0] = 300
print(f'Vendas do produto {produtos[0]} foram de {vendas[0]} unidades')
```

```python
texto = 'lira@gmail.com'
print(texto)
```

```python
#texto[1] = 'a' #string não é multável
#print(texto) -> 'str' object does not support item assignment
novo_texto = texto.replace('i', 'a') #escrevendo uma nova variável
print(novo_texto)
texto = texto.replace('i', 'a') #sobreescrevendo a variável texto
print(texto)
```

###Como descobrir o índice de um item de uma lista

i = lista.index('item')

```python
products = ['tv', 'celular', 'tablet', 'mouse', 'teclado', 'geladeira', 'forno']
estoque = [100, 150, 100, 120, 70, 180, 80]
```

```python
i = products.index('geladeira')
qtde_estoque = estoque[i]
print(f'Quantidade em estoque de geladeira é de :{qtde_estoque}')
```

```python
#Crie um programa para fazer uma consulta de estoque. O usuário do programa deve inserir o nome do produto e, caso ele não exista na lista, ele é avisado. Caso exista, o programa deve dizer a quantidade de unidades em estoque do produto.

produto = input('Insisra o nome do produto: ').lower()
if produto in products:
    item = products.index(produto)
    print(f'O produto {produto} possui {estoque[item]} unidades')
else:
    print(f'O produto {produto} não está na lista de produtos da loja')
```

###Algumas funções Básicas de lista

- Tamanho da lista -> tamanho = len(lista)
- Maior ou menor valor -> maior = max(lista) e menor = min(lista)

```python
tamanho = len(product)
tamanho
```

```python
vendas = [1000, 1500, 15000, 270, 900, 100, 1200]

#Qual o item mais vendidos?
mais_vendidos = max(vendas)

#Qual o item menos vendidos?
menos_vendidos = min(vendas)

print(f'O produto mais vendido foi {mais_vendidos} e o menor valor foi de {menos_vendidos}')
```
