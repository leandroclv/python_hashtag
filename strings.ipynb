###'Formulas' de texto - Métodos de string

####Estrutura:
- Normalmente usamos a estrutura texto.método() para fazer as modificações que queremos 
- Alguns métodos pedem 'argumentos', que são informações que temos que passar para a fórmula (método) para ela funcionar. Esses argumentos são passados dentro do parênteses

####Como usar:

Não decore os métodos, os que você for mais usando com o tempo você vai decorar o que precisa

Mas a dica é: use essa lista para consulta e busque entender como os métodos funcionam e suas aplicações, para poderconsultar e usar quando precisar.

    - capitalize() -> Coloca a 1° letra maiúscula
    - ex: texto = 'lira'
    -     print(exto.capitalize()) -> Lira
    - casefold() -> Transforma todas as letras em minúscula (existe lower() mas o casefold é melhor normalmente)
    - ex: texto = 'Lira'
    -     print(texto.casefold()) -> lira
    - count() -> Quantidade de vezes que um valor aparece na string
    - ex: text = 'lira@yahoo.com.br'
    -     print(texto.count('.')) -> 2
    - endswith() -> Verifica se o texto termina com um valor específico e dá como resposta True ou False
    - ex: texto = 'lira@gmail.com'
    -     print(texto.endswith('gmail.com')) -> True
    - find() -> Procura um texto dentro de outro texto e dá como resposta a posição do texto encontrado
    - ex: texto = 'lira@gmail.com'
    -     print(texto.find('@')) -> 4
    - format() -> Formata uma string de acordo com os valores passados. Já usamos bastante ao long do programa
    - ex: texto = 'Lira'
    -     print('Olá {}, como vai?'.format(texto)) -> Olá Lira, como vai?
    - isalnum() -> Verifica se um texto é todo feito com caracteres alfanuméricos (letras e números) -> letras com acentos ou ç são considerados letras sem função
    - ex: texto = 'João123'
    -     print(texto.isalnum()) -> True
    -     OBS: se o texto fosse 'Jo~ao' ou 'Joao#' o esultado seria false
    - isalpha() -> Verifica se um texto é todo feito de letras
    - ex: texto = 'João'
    -     print(texto.isalpha()) -> True
    -     OBS: nesse caso se o texto fosse 'Joao123' o resultado seria False, porque 123 não são letras
    - isnumeric() -> Verifica se um texto é todo feito por números
    - ex: texto = '123'
    -     print(texto.isnumeric()) -> True
    -     OBS: existem os métodos isdigit() e isdeciaml() que tem variações pontuais em caracteres especiais tipo textos com frações e potências, mas para 99% dos casos eles não vão ser necessários
    - replace() -> Substitui um texto porum outro texto em uma string
    - ex: texto = '1000.00'
    -     print(texto.replace('.', ',')) -> 1000,00
    -     OBS: O replace precisa de dois argumentos para funcionar.O 1° é o texto que vocêquer trocar. O 2° é o texto que você quer colocar no lugar daquele texto que você esta tirando
    - split() -> Separa uma string de acordo com um delimitador em vários textos diferentes
    - ex: texto = 'lira@gmail.com'
    -     print(texto.split('@')) -> ['lira', '@gmail.com']
    - splitlines() -> separa um texto em vários textos de acordo com os 'enters' do texto
    - ex: texto = '''
    - Olá, bom dia 
    - Venho por meio desse email lhe informar o faturamento de hoje.
    - Faturamento = R$2.500,00 
    - '''
    -     print(texto.splitlines()) -> ['Olá, bom dia', 'Venho por meio desse email lhe informar o faturamento de hoje.', 'Faturamento = R$2.500,00']
    - startswith() -> Verifica se a string começa com determinado texto
    - ex: texto = 'BEB123453'
    -     print(texto.startswith('BEB')) -> True
    - strip() -> retira caracteres indesejados dos textos. Por padrão, retira espaços 'extras' no iníicio e no final
    - ex: texto = ' BEB123453 '
    -     print(texto.strip()) -> 'BEB123453'
    - title() -> Coloca a 1° letra de cada palavra em maiúscula
    - ex: texto = 'joão paulo lira'
    -     print(texto.title()) -> João Paulo Lira
    - upper() -> Coloca o texto todo em letra maiúscula
    - ex: texto = 'beb123453'
    -     print(texto.upper()) -> BEB123453

###Format - Aula de consulta

Como usar o format para criar formatações personalizadas em prints e textos
- :< -> Alinha o texto à esquerda (se tiver espaço na tela para isso)
- :> -> Alinha o texto à direita (se tiver espaço na tela para isso)
- :^ -> Alinha o texto ao centro (se tiver espaço na tela para isso)
- :+ -> Coloca o sinal sempre na frente do número (independente se é positivo ou negativo)
- :, -> Coloca a virgula como separador de milhar
- :_ -> Coloca o _ como separador de milhar
- :e -> Formato científico
- :f -> Número com quantidade fixade casas decimais  
- :x -> Formato HEX minúscula (para cores)
- :X -> Formato HEX maiúscula (para cores)
- :% -> Formato percentual

```python
#Alinhamento
email = 'lira@gmail.com'
print(f'Meu email não é {email:<30}, show?')
print(f'Meu email não é {email:^30}, show?')
```

```python
#Edição de sinal
custo = 500
faturamento = 270
lucro = faturamento - custo
print(f'Faturamento foi {faturamento:+} e lucro foi {lucro:+}')
```

```python
#Separador de milhar
custo = 5000
faturamento = 2700
lucro = faturamento - custo
print(f'Faturamento foi {faturamento:,} e lucro {lucro:,}')
```

```python
#Formato com casas decimais fixas
custo = 500
faturamento = 270
lucro = faturamento - custo
print(f'Faturamento foi {faturamento:.2f} e lucro {lucro:.2f}')
```

```python
#Formato percentual
custo = 500
faturamento = 1300
lucro = faturamento - custo
margem = lucro / faturamento
print(f'Margem de lucro foi de: {margem:.1%}')
```

```python
#Formato moeda -> combinação de formatos
#Existem módulos/bibliotecas que vão facilitar isso, caso a gente queira, mas vamos ver como usar módulos mais a frente do curso. Por enquanto, se você precisar, pode fazer substituição em string

custo = 5000
faturamento = 27000
lucro = faturamento - custo
print(f'Faturamento foi R${faturamento:,.2f} e lucro R${lucro:,.2f}')

#Transformando no fortmato brasileiro
lucro_texto = f'R${lucro:_.2f}'
print(lucro_texto.replace('.', ',').replace('_', '.'))
```

####1.Cadastro de CPF

Crie um programa para cadastro de CPF de clientes que recebe o CPF em um input box apenas com números

Ex: 'Insira seu CPF (digite apenas números)'

Caso o usuário digite algo diferente de números ou digite menos de 11 caracteres (tamanho do CPF brasileiro), o programa deve exibir uma mensagem de 'Digite seu CPF corretamente e digite apenas números'

```python
cpf = input('Insira o seu CPF (digite apenas números): ')

if len(cpf) <= 0:
    print('Você não digitou nada')
elif not cpf.isnumeric():
    print('Digite somente números no seu CPF')
elif not len(cpf) == 11:
    print('Digite a quantidade certa')
else:
    print('CPF inserido corretamente')
```

####2. Melhorando nosso cadastro de CPF

Agora, além das validações anteriores, vamos criar um input que permita que o usuário insira pontos, traços e inclusive espaços vazios.

Nosso programa deve 'tratar' o que o usuário inserir para padronizar o CPF dele em apenas números

A verificação de tamanho do CPF com 11 caracteres continua válida, mas ela só deve ser feita depois de retirar todos os pontos, traços e espaços do CPF que o cliente inserir e, uma vez retirados pontos,traços e espaços, devem sobrar apenas números no CPF. Qualquer outro caractere deve ser considerado inválido

No final, nosso programa deve exibir uma mensagem para o usuário, caso ele tenha inserido o CPF inválido ou então apenas deve printar o CPF só com número

```python
cpf = input('Insira o seu CPF: ')

#tratar o cpf
cpf_tratado = cpf.strip().replace('.','').replace('-','')


if len(cpf_tratado) == 11 and cpf_tratado.isnumeric() and cpf_tratado:
    print(cpf_tratado)
else:
    print('Digite seu cpf corretamente e digite apenas números')
```

###3.Cadastro de e-mails 

- A Hashtag sempre se comunica com seus clientes por e-amil. Para isso, a gente tem em cada página um cadastro de nome e e-mail. Nesse cadastro, nosso sistema verifica se o e-mail que a pessoa inseriu é um e-mail válido, verificando se ele tem '@' e se depois do '@' tem algum ponto, afinal:
    - liragemail.com NÃO é um e-mail válido
    - lira@gmail NÃO é um e-mail válido
    - lira@gmail.com é um e-mail válido

Crie um programa que permita o cadastro de nome e e-mail de uma pessoa (por meio de inputs) e que verifique:
    - 1. Se o nome e e-mail foram preenchidos, caso contrário ele deve avisar para preencher todos os dados corretamente
    - 2. Se o e-mail contém '@' e se depois do '@' existe algum '.', caso contrário ele deve exibir uma mensagem de e-mail inválido

OBS: Pode te ajudar lembrar do método .find da aula de métodos de string. Você pode testar o que ele dá como resposta caso ele não encontre um item dentro da string

```python
nome = input('Digite seu nome: ')
email = input('Digite o seu e-mail: ')

if nome and email:
    pos_a = email.find('@')
    servidor = email[pos_a:]
    if pos_a != -1 and '.' in servidor:
        print('cadastro concluido')
    else:
        print('e-mail inválido')
else:
    print('Digite o seu nome e e-mail corretamente')
```
