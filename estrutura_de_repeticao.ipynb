```python
vendas = [1200, 300, 800, 1500, 1900, 2750, 400, 20, 23, 70, 90, 80, 1100, 999, 900, 880, 870, 50, 1111, 120, 300, 450, 800]
meta = 1000
qtd_bateu_meta = 0

for venda in vendas:
    if venda >= meta:
        #print(venda, end=' ')
        qtd_bateu_meta += 1

print(f'O percentual de funcionários que bateram a meta foi de {(qtd_bateu_meta / len(vendas)) * 100:.1f}%')
```

###Enumerate

- Estrutura:
    - O enumerate permite que você percorra uma lista e ao mesmo tempo tenha em uma variável o índice daquele item

```python
#for item in lista:
#   resto do código

#for i, item in enumerate(lista):
#   resto do código
```

```python
funcionarios = ['Maria', 'José', 'Antônio', 'João', 'Francisco', 'Ana', 'Luiz', 'Paulo', 'Carlos', 'Manoel', 'Pedro', 'Francisca', 'Marcos', 'Raimundo', 'Sebastião', 'Antônia', 'Marcelo', 'Jorge', 'Marcia', 'Geraldo', 'Adriana', 'Sandra', 'Luis']

for i, funcionario in enumerate(funcionarios):
    print(f'O indice {i} é o funcionário {funcionario}')
```

```python
#Exemplo prático
#Vamos pegar um exemplo denível mínimo de estoque. Em uma fábrica você tem vários produtos e não pode deixar que os produtos fiquem em falta. Para isso, foi definidouma quantidade mínima de estoque que os produtos precisam ter:

#Identifique quais produtos estão abaixo do nível mínimo de estoque:

estoque = [1200, 300, 800, 1500, 1900, 2750, 400, 20, 23, 70, 90, 80, 1100, 999, 900, 880, 870, 50, 1111, 120, 300, 450, 900]
produtos = ['coca', 'pepsi', 'guarana', 'skol', 'brahma', 'água', 'del valle', 'dolly', 'red bull', 'cachaça', 'vinho tinto']
nivel_minimo = 50

print(len(estoque))
print(len(produtos))
```

```python
for i,qtde in enumerate(estoque):
    if qtde < nivel_minimo:
        print(f'{produtos[i]} está abaixo do nível mínimo. Temos apenas {qtde} unidades')
```

```python
produtos = ['iphone', 'ipad', 'airpod','macbook']
precos = [7000, 10000, 2500, 14000]
```

```python
#For item in lista
#preço com imposto
for preco in precos:
    print(preco * 1.1)
```

```python
#For in range
#preço de cada produto
for i in range(len(precos)):
    produto = produtos[i]
    preco = precos[i]
    print(produto, preco)
```

```python
#For item in lista com enumerate
#preço de cada produto com imposto
for i, preco in enumerate(precos):
    preco *= 1.1
    produto = produtos[i]
    print(produto, preco)
```

###For dentro de for

Quando temos listas dentro de listas, às vezes precisamos fazer um 'for dentro de for'

Vamos pegar um exemplo de nível mínimo de estoque. Em uma fábrica você tem vários produtos e não pode deixar que os produtos fiquem em falta. Para isso, foi definido uma quantidade mínima de estoque que os produtos precisam ter:

Indique quais fábricas tem algum produto abaixo do nível de estoque

Agora ao invés de analisar o estoque de apenas 1 fábrica, vamos analisar o estoque de várias fábricas

```python
estoque = [
        [294, 125, 269, 208, 783, 852, 259, 371, 47, 102, 386, 87, 685, 686, 697, 941, 163, 631, 7, 714, 218, 670, 453],
        [648, 816, 310, 555, 992, 643, 226, 319, 501, 23, 239, 42, 372, 441, 126, 645, 927, 911, 761, 445, 974, 2, 549],
        [832, 683, 784, 449, 977, 705, 198, 937, 729, 327, 339, 10, 975, 310, 95, 689, 137, 795, 211, 538, 933, 751, 522],
        [837, 168, 570, 397, 53, 297, 966, 714, 72, 737, 259, 629, 625, 469, 922, 305, 782, 243, 841, 848, 372, 621, 362],
        [429, 242, 53, 985, 406, 186, 198, 50, 501, 870, 781, 632, 781, 105, 644, 509, 401, 88, 961, 765, 422, 340, 654],
    ]
fabricas = ['Lira Manufacturing', 'Fábrica Hashtag', 'Python Manufaturas', 'Produções e Cia', 'Manufatura e Cia']
nivel_minimo = 50
fabricas_abaixo_nivel = []
```

```python
for i, lista in enumerate(estoque):
    #se dentro daquela lista tem alguém abaixo do estoque
    for qtde in lista:
        if qtde < nivel_minimo:
            if fabricas[i] in fabricas_abaixo_nivel:
                pass
            else:
                fabricas_abaixo_nivel.append(fabricas[i])
                
print(fabricas_abaixo_nivel)
```
