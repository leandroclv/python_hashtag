###Tuplas

- Estrutura:
    - tupla = (valor, valor, valor,...)

- Diferença
    -Parece uma lista, mas é imutável

- Vantagens
    - Mais eficiente (em termo de performance)
    - Protege a base de dados (por ser imutável)
    - Muito usado para dados heterogênios

```python
#Criando tuplas
vendas = ('Lira', '25/08/2020', '15/02/1994', 2000, 'Estagiário')
```

```python
#Acessando o valor de uma tupla
nome = vendas[0]
data_contratação = vendas[1]
data_nascimento = vendas[2]
salario = vendas[3]
cargo = vendas[4]


nome, data_contratação, data_nascimento, salario, cargo = vendas #Mesma coisa que o que é feito acima
```

```python
print(nome)
cargo
```
