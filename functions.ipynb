###Function Python

As functions são blocos de código que servem 1 único propósito, fazem uma ação específica

```python
#Estrutura básica
#def nome_funcao():
#   faça alguma coisa
#   faça outra coisa
#   return valor_final
```

```python
#Exemplo: vamos criar uma função de cadastro de um produto. Essa função deve garantir que o produto cadastrado está em letra minúscula

def cadastrar_produto():
    produto = input('Digite o nome doproduto que deseja cadastrar: ')
    produto = produto.casefold()
    print(f'Produto {produto} cadastrado com sucesso')
```

```python
cadastrar_produto()
```

```python
for i in range(3):
    cadastrar_produto()
```

###Retornar um valor na Function Python

```python
#Estrutura básica
#def nome_funcao():
#   return valor_final
```

```python
#Ex: Vamos criar uma função de cadastro de um Produto. Essa função deve garantir que o produto cadastrado está em letra minúscula
def cadastrar_produto():
    produto = input('Digite o nome doproduto que deseja cadastrar: ')
    produto = produto.casefold()
    produto = produto.strip()
    return produto
```

```python
variavel_produto = cadastrar_produto()

print(variavel_produto)
```
