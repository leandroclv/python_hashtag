####time.strftime()

A função strftime() converte um objeto de tempo struct para uma string de acordo com um formato específico

Os símbolos de formato que podem ser usados estão disponíveis na documentação oficial do Python

Por exemplo, podemos querer uma string de tempo no seguinte formato: 'Dia da semana, dia do mês de mês do ano, horas:minutos:segundos'. Podemos usar o seguinte código para obter o tempo formatado:

```python
tempo_em_struct = time.localtime()

print(tempo_em_struct)

print(time.strftime('%d %B %Y', tempo_em_struct)) 
#%d -> dia do mês como número decimal; %B -> nome completo do mês da localidade; %Y -> ano com século como número decimal
print(time.strftime('%H:%M:%S', tempo_em_struct))
#%H -> hora (relógio 24h) como número decimal; %M -> minuto como número decimal; %S -> segundo como número decimal
```

```python
tempo_formatado = time.strftime('%A, %d de %B de %Y, %H:%M:%S', tempo_em_struct)
print(f'Tempo formatado: {tempo_formatado}')
```

```python
#Observe que o dia e o mês aparecem em inglês. Para obter o tmepo formatado em português podemos usar o módulo locale do python
import locale

#definir a localização para português
locale.setlocale(locale.LC_TIME, 'pt_BR.UTF-8')

tempo_em_struct = time.localtime()
tempo_formatado = time.strftime('%A, %d de %B de %Y, %H:%M:%S', tempo_em_struct)
print(f'Tempo formatado: {tempo_formatado}')
```

####time.strptime()

A função strptime() analisa uma string representando um horário de acordo com um formato. O retorno é um objeto struct_time

```python
string_tempo = '30 Junho, 2026'
formato = '%d %B, %Y'
tempo_em_struct = time.strptime(string_tempo, formato)

print(f'Tempo em struct: {tempo_em_struct}')
```

```python
#data em formato dia/mês/ano
string_tempo = '06/09/2023'
formato = '%d/%m/%Y'
tempo_em_struct = time.strptime(string_tempo, formato)
print(tempo_em_struct)
```

```python
#data em formato mês/dia/ano
string_tempo = '06/09/2023'
formato = '%m/%d/%Y'
tempo_em_struct = time.strptime(string_tempo, formato)
print(tempo_em_struct)
```

####time.gmtime()

A função gmtime() converte um tempo expresso em segundos desde a EPOCH em um objeto struct_time em UTC. UTC significa Coordinated Universal Time, também conhecido como GMT (Greenwich Mean Time). Este é o fuso horário padrão em que as funções time operam

```python
#Tempo em UTC para o segundos desde o EPOCH
time.gmtime(0)
```

```python
gmt_struct = time.gmtime() #tempo atual em UTC já que não fornecemos nenhum argumento
print(f'Tempo em UTC: {gmt_struct}')
```

```python
#comparando com Localtime
print(f'Tempo local: {time.localtime()}')
print(f'Tempo em UTC: {time.strftime('%A, %d de %B de %Y, %H:%M:%S', gmt_struct)}')
print(gmt_struct.tm_zone)
```

```python
gmt_struct_exemplo = time.gmtime(1_234_567_890)
print(f'Tempo em UTC: {gmt_struct_exemplo}')
print(f'Tempo em UTC: {time.strftime('%A, %d de %B de %Y, %H:%M:%S', gmt_struct_exemplo)}')
```

####time.mktime()

A função mktime() converte um objeto struct_time em segundos desde a EPOCH. Este é o inverso da função localtime(). Por exemplo,podemos converter o objeto struct_time retornado pela função localtime() em segundos desde a EPOCH usando a função mktime(). O resultado deve ser o mesmo que o valor retornado pela função time()

```python
tempo_em_struct = time.localtime()
tmepo_em_segundos = time.mktime(tempo_em_struct)
print(f'Tempo em segundos: {tmepo_em_segundos}')
print(f'Tempo em segundos: {time.time()}')
```

Podemos usar o método mktime para calcular a diferença entre dois tempos. Por exemplo, podemos calcular a diferença entre o tempo atual e o início do ano. O resultado será o número de segundos entre esses dois tempos. Podemos usar a função localtime() para obter o tempo atual e a função mktime() para obter o tepmo em 01 de janeiro de 2023 (ano deste material).

```python
tempo_atual = time.localtime()
tempo_ano_novo = time.mktime((2023, 1, 1, 0,0,0,0,0,0))

diferenca = time.mktime(tempo_atual) - tempo_ano_novo
print(f'Diferença em segundos: {diferenca}')
```

###Criação de uma data a partir dos valores

Podemos criar um objeto datetime usando o datetime. Você pode passar:

- year: ano (por exemplo,2023)
- month: mês (1-12)
- day: dia (1-31)
- hour: hora (0-23)
- minute: minuto (0-59)
- second: segundo (0-59)
- microsecond: microssegundo (0-999999)
- tzinfo: fuso horário

```python

from datetime import datetime, timedelta

dia = 10
mes = 2
ano = 1997

data_analise = datetime(ano, mes, dia)
data_analise
```

####datetime.timedelta()

A classe timedelta é usada para realizar operações com datas (adição e subtração)

```python
data_inicio = datetime(2026, 2, 15)
duracao = 387
data_final = data_inicio + timedelta(days=duracao)
data_final
```

```python
duracao_semanas = 15
data_final = data_inicio + timedelta(weeks=duracao_semanas)
print(data_final)
```

###Calcular a diferença entre duas datas

Podemos calcular a diferença entre duas datas subtraindo uma da outra. O resultado será um objeto timedelta

```python
tempo_projeto = data_final - data_inicio
print(tempo_projeto.days)
```

```python
nova_data_projeto = datetime(2026, 4, 15)
nova_data_fim = nova_data_projeto + tempo_projeto
print(nova_data_fim.day)
```

###Guia para a biblioteca datetime em Python

####datetime.datetime.strftime()

A função strftime() converte um objeto datetime para uma string de acordo com um formato específico.

```python
from datetime import datetime, timezone, timedelta
from zoneinfo import ZoneInfo, available_timezones
import locale

locale.setlocale(locale.LC_TIME, 'pt_BR.UTF-8')


data_inicio = datetime(2026, 2, 15)
data_inicio_texto = data_inicio.strftime('%d %B %Y, %A')
print(data_inicio_texto)
```

####datetime.datetime.strptime()

A função strptime() analisa uma string representando uma data e hora de acordo com um formato. O retorno é um objeto datetime

```python
data_inicio_texto = '15 de fevereiro de 2026'
data_inico = datetime.strptime(data_inicio_texto, '%d de %B de %Y')
print(data_inico)
```

```python
#Trabalhando com fuso horário
fuso_horario = timezone.utc
data_inicio = datetime(2026, 2, 15, 15, 25, tzinfo=fuso_horario)
print(data_inicio)
```

```python
fuso_horario = timezone(timedelta(hours=-3))
data_inicio = datetime(2026, 2, 15, 15, 25, tzinfo=fuso_horario)
print(data_inicio)
```

```python
fuso_horario_NY = timezone(timedelta(hours=-4))
data_inicial_NY = data_inicio.astimezone(fuso_horario_NY)
print(data_inicial_NY)
```

```python
novo_inicio_fuso_ny = data_inicial_NY + timedelta(hours=2)
print(novo_inicio_fuso_ny)
print(novo_inicio_fuso_ny.astimezone(fuso_horario))
```

```python
fuso_sp = ZoneInfo('America/Sao_Paulo')
data_inicio = datetime(2026, 2, 15, 15, 25, tzinfo=fuso_sp)
print(data_inicio)
```

```python
fuso_ny = ZoneInfo('America/New_York')
print(data_inicio.astimezone(fuso_ny))
```
